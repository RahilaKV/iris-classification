```python
x=df.iloc[:,1:5]
x
```

```python
y=df["Species"]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
x_m=m.fit_transform(x)
x_m
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_m,y,test_size=0.3,random_state=1)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier()
k.fit(x_train,y_train)
```

```python
y_k=k.predict(x_test) #y_test predicted
y_k 
```

```python
y_test #actual
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(ConfusionMatrixDisplay.from_predictions(y_test,y_k))
print(classification_report(y_test,y_k))
```

### Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
p={'n_neighbors':[1,3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
t=GridSearchCV(k,p,cv=5,scoring='accuracy')
t.fit(x_train,y_train)
print(t.best_params_)
```

```python
kn=KNeighborsClassifier(algorithm='auto',n_neighbors=5,weights='distance')
kn.fit(x_train,y_train)
```

```python
y_kn=kn.predict(x_test)
y_kn
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_kn))
print(classification_report(y_test,y_kn))
```
